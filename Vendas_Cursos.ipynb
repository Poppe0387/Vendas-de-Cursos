Nesta seção, os dados são carregados e pré-processados. Incluímos etapas de verificação de valores nulos,

```python
# Análise específica de vendas, sazonalidade e desempenho
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Análise específica de vendas, sazonalidade e desempenho
#Importando o arquivo .csv e exibindo as primeiras linhas
vendas = pd.read_csv("VendasDeCursos.csv")
vendas.head(5)
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# transformando a primeira coluna "ID" em index para ficar mais simples a visualização
vendas = vendas.set_index('ID')
vendas.head(5)
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Exibindo as informações básicas sobre o dataset
vendas.info()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Verificando se existem valores nulos
vendas.isnull().sum()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Calculando as estatísticas descritivas para colunas numéricas
vendas.describe()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Descobrindo a quantidade de vendas por curso com gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(x='Quantidade de Vendas', y='Nome do Curso', data = vendas)
plt.title('Quantidade de Vendas por curso')
plt.xlabel('Quantidade de Vendas')
plt.ylabel('Nome do Curso')
plt.show()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Explorando a relação entre preço unitário e quantidade de vendas a partir do gráfico de dispersão
plt.figure(figsize=(10,6))
sns.scatterplot(x='Preço Unitário', y='Quantidade de Vendas', data = vendas)
plt.title('Relação entre Preço Unitário e Quantidade de Vendas')
plt.xlabel('Preço Unitário')
plt.ylabel('Quantidade de Vendas')
plt.show()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Calculando a receita total gerada pela venda dos cursos
vendas['Receita'] = vendas['Quantidade de Vendas'] * vendas['Preço Unitário']
receita_total = vendas['Receita'].sum()

print(f"Receita total gerada: R$ {receita_total:.2f}.")
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Identificando o curso com o maior número de vendas
curso_mais_vendido = vendas.loc[vendas['Quantidade de Vendas'].idxmax()]

print(f"Curso com o maior número de vendas foi {curso_mais_vendido['Nome do Curso']}.")
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Visualizando a distribuição das vendas ao longo do tempo
# Primeiramente convertendo a coluna 'Data' para formato datetime
vendas['Data'] = pd.to_datetime(vendas['Data'])

plt.figure(figsize=(10,6))
vendas.groupby('Data')['Quantidade de Vendas'].sum().plot(kind='line')
plt.title('Distribuição das Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.grid(True)
plt.show()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Descobrindo quais os cursos mais rentáveis 

receita_por_curso = vendas.groupby('Nome do Curso')['Receita'].sum().sort_values(ascending=False)
receita_por_curso
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Explorando a distribuição de preços dos cursos
plt.figure(figsize=(10,6))
sns.histplot(vendas['Preço Unitário'], bins=10, kde=True)
plt.title('Distribuição dos Preços dos Cursos')
plt.xlabel('Preço Unitário')
plt.ylabel('Frequência')
plt.show()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Realizando a análise de correlação entre preço e quantidade de vendas para descobrir se preços mais baixos impulsionam vendas mais altas ou se cursos mais caros vendem mais devido ao seu valor percebido.
correlacao = vendas[['Preço Unitário', 'Quantidade de Vendas']].corr()
print(correlacao)

# Visualizando a correlação
plt.figure(figsize=(10,6))
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title('Correlação entre Preço Unitário e Quantidade de Vendas')
plt.show()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Buscando outliers que identifiquem se existem cursos que estão fora do padrão esperado de vendas ou preço, através de boxplot.
# Boxplot para verificar outliers nas vendas
plt.figure(figsize=(10,6))
sns.boxplot(x=vendas['Quantidade de Vendas'])
plt.title('Distribuição das Vendas (Outliers)')
plt.show()

# Boxplot para verificar outliers nos preços
plt.figure(figsize=(10,6))
sns.boxplot(x=vendas['Preço Unitário'])
plt.title('Distribuição dos Preços (Outliers)')
plt.show()
```

```python
# Análise específica de vendas, sazonalidade e desempenho
# Analisando a relação entre receita e vendas para verificar se os cursos com preços mais altos geram receitas proporcionais. Calculando para isso a receita gerada por cada curso.
# Receita por venda
vendas['Receita por Venda'] = vendas['Receita'] / vendas['Quantidade de Vendas']

# Visualizando a relação entre preço unitário e receita por venda
plt.figure(figsize=(10,6))
sns.scatterplot(x='Preço Unitário', y='Receita por Venda', data=vendas)
plt.title('Relação entre Preço Unitário e Receita por Venda')
plt.xlabel('Preço Unitário')
plt.ylabel('Receita por Venda')
plt.show()
```
